Stemming:

Stemming is the process of reducing a word to its Root word

example: actor, actress, acting --> act

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
covid_data['content'] = covid_data['content'].apply(stemming)
```

```python
print(covid_data['content'])
```

```python
X = covid_data['content'].values
Y = covid_data['type'].values
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
print(X)
```

Spliting the dataset to training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the Model: LogisticRegression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Evaluation: Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('the accuracy of training data is :', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('the accuracy of the testing data is:', testing_data_accuracy)
```
